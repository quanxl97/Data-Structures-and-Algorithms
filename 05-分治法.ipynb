# 分治法
Divide-and-Conquer

一种经典优化算法

【基本介绍】  
直观理解：将一个难以解决的大问题分割成一些规模比较小的相同的问题进行求解。  
用到了递归，自己调用自己  

【栗子】  
100枚硬币，寻找假币，假币略轻  
传统做法，逐个硬币进行比较  

分治法：  
（1）将100枚硬币分成3份，33，33，34  
（2）比较33，33两份，若相等，则假币在34中，若不相等，在轻的那一份中  
（3）已经确定假币在哪一份中，假如在33那份中，将33再分为11，11，11  
（4）重复步骤2，确定假币在其中一份11中  
（5）将11分成4，4，3  
（6）继续重复步骤2，假设假币在其中一份4中  
（7）将4分成1，1，2，重复步骤2，假设假币在2中  
（8）比较2的两枚硬币，轻的那个是假币，假币找到。  
这样，比较了5次就可以解决问题，大大的提高了效率。

【分治法流程】  
划分问题：整个问题划分成多个无关联的子问题   
递归求解：递归调用求解各个子问题  
合并问题：合并子问题的解，形成原始问题的解。  

【使用条件】  
1·该问题规模缩小到一定程度后可以容易解决  
2·该问题可以分解为若干个规模较小的相同问题  
3·利用该问题分解出的子问题的解合并后可以获得原始问题的解  
4·该问题所分解出的各个子问题是相互独立的，即子问题之间不包含公共的子子问题（非必须条件）  

使用条件分析：  
条件2，3是分治的前提（必要条件）  
条件4，对于存在公共子问题的问题，使用分治法会存在重复计算的问题，使用动态规划较为合适。  

【分治法与动态规划】  
分治与动态规划有共通也有不同之处  
当子问题相互独立时，能且只能使用分治，，存在重叠子问题时，动态规划是更好的算法。  
小结：分治法——各个子问题独立，动态规划——各子问题重合
